```python
pbr_list = []
for i in range(0, len(kospi_bottom)):
    try:
        company_name = kospi_bottom[i]
        company_ifrs = pd.read_excel('/workspace/Jupyter_notebook/financial_analysis/data/ifrs/'+company_name + '.xlsx')
        pbr_recently = len(company_ifrs['PBR(배)'].dropna()) - 1
        pbr = company_ifrs['PBR(배)'][pbr_recently]
        
        if(pbr < 0.2):
            continue
        else:
            pbr_list.append([kospi_bottom[i], pbr])
    
    except:
        print('ETF:'+ kospi_bottom[i])
```

```python
pbr_df = pd.DataFrame(pbr_list, columns= ['기업명', 'PBR'])
```
