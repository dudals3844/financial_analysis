```python
import requests
import urllib.request
import pandas as pd
from bs4 import BeautifulSoup 
from pandas import ExcelWriter
import xlsxwriter
from datetime import datetime
import pandas_datareader.data as wb
```

```python
kosdaq = pd.read_excel('/workspace/Jupyter_notebook/financial_analysis/data/kosdaq.xls')
```

```python
stock_code_list = []
def two_len(code):
    code = '0000' + code
    return code

def three_len(code):
    code = '000' + code
    return code

def four_len(code):
    code = '00' + code
    return code
    
def five_len(code):
    code = '0' + code
    return code

for i in range(0, len(kosdaq['종목코드'])):
    if(len(str(kosdaq['종목코드'][i])) == 4):
        stock_code_list.append([four_len(str(kosdaq['종목코드'][i])), kosdaq['기업명'][i]])
    elif(len(str(kosdaq['종목코드'][i])) == 5):
        stock_code_list.append([five_len(str(kosdaq['종목코드'][i])), kosdaq['기업명'][i]])
    elif(len(str(kosdaq['종목코드'][i])) == 3):
        stock_code_list.append([three_len(str(kosdaq['종목코드'][i])), kosdaq['기업명'][i]])
    elif(len(str(kosdaq['종목코드'][i])) == 2):
        stock_code_list.append([two_len(str(kosdaq['종목코드'][i])), kosdaq['기업명'][i]])
    else:
        stock_code_list.append([str(kosdaq['종목코드'][i]), kosdaq['기업명'][i]])
    
    
```

```python
total_code = []
for i in range(len(stock_code_list)):
    code = 'A' + stock_code_list[i][0]
    total_code.append(code)
```

```python
def investment_indicator_kosdaq(code):
    url = 'http://comp.fnguide.com/SVO2/ASP/SVD_Invest.asp?pGB=1&gicode=' + code + '&cID=&MenuYn=Y&ReportGB=&NewMenuID=105&stkGb=701'
    html = urllib.request.urlopen(url).read()
    soup = BeautifulSoup(html, 'html.parser')
    table = soup.find('table', {'class':'us_table_ty1 h_fix zigbg_no'})
    investment_indicator = pd.read_html(str(table))
    investment_trans = investment_indicator[0].transpose().reset_index()
    return investment_trans
```

```python
def make_excel(df, company):
    file_name = '/workspace/Jupyter_notebook/financial_analysis/data/investment_indicator_kosdaq/'+company+'.xlsx'
    writer = pd.ExcelWriter(file_name, engine='xlsxwriter')
    df.to_excel(writer, sheet_name='company_list')
    writer.close()
```

```python
for i in range(len(total_code)):
    try:

        df = investment_indicator_kosdaq(total_code[i])
        make_excel(df, kosdaq['기업명'][i])
    except :
        print('Error')
```
