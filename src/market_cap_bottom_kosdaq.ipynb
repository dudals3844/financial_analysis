```python
import requests
import pandas as pd
from bs4 import BeautifulSoup as bs
from pandas import ExcelWriter
import xlsxwriter
from datetime import datetime
import pandas_datareader.data as wb
```

```python
kosdaq_market_cap = pd.read_excel('/workspace/Jupyter_notebook/financial_analysis/data/kosdaq_market_cap.xls')
```

```python
kosdaq_market_cap
```

```python
start = len(kosdaq_market_cap['종목명']) - 600 - 1
end = len(kosdaq_market_cap['종목명']) - 1
kosdaq_bottom = []
```

```python
for i in range(start, end):
    kosdaq_bottom.append(kosdaq_market_cap['종목명'][i])
```

```python
start
```

```python
pbr_list = []
for i in range(0, len(kosdaq_bottom)):
    try:
        company_name = kosdaq_bottom[i]
        company_ifrs = pd.read_excel('/workspace/Jupyter_notebook/financial_analysis/data/ifrs_kosdaq_quarterly/'+company_name + '.xlsx')
        pbr_recently = len(company_ifrs['PBR(배)'].dropna()) - 1
        company_ifrs['PBR(배)'].dropna(inplace = True)
        company_pbr = company_ifrs['PBR(배)'].reset_index()
        pbr = company_pbr['PBR(배)'][pbr_recently]
        
        
        if(pbr < 0.2):
            continue
        else:
            pbr_list.append([kosdaq_bottom[i], pbr])
    
    except:
        print('ETF:'+ kosdaq_bottom[i])
```

```python
pbr_df = pd.DataFrame(pbr_list, columns= ['기업명', 'PBR'])
```

```python
pbr_df
```

```python
pbr_df.sort_values(by=['PBR'],ascending=True).dropna()
```

```python
per_list = []
for i in range(0, len(kosdaq_bottom)):
    try:
        company_name = kosdaq_bottom[i]
        company_ifrs = pd.read_excel('/workspace/Jupyter_notebook/financial_analysis/data/ifrs_kosdaq_quarterly/'+company_name + '.xlsx')
        per_recently = len(company_ifrs['PER(배)'].dropna()) - 1
        company_ifrs['PER(배)'].dropna(inplace = True)
        company_per = company_ifrs['PER(배)'].reset_index()
        #print(company_per)
        per = company_per['PER(배)'][per_recently]
        if(per < 0):
            continue
        elif(per < 10):
            per_list.append([kosdaq_bottom[i], per])

            
    except:
        print()
```

```python
per_df = pd.DataFrame(per_list, columns= ['기업명', 'PER'])
```

```python
per_df.sort_values(by=['PER'],ascending=True).dropna().reset_index()
```
